```python
df1=pd.read_csv("problems.csv")
```

```python
df1.head()
```

```python
topics=df1["Topic"].unique()
print(topics)
```

```python
tp=df1["Topic"].value_counts()
print(tp)
```

```python
print(tp["Tree"])
```

```python
#pie chart for problem distribution

import matplotlib.pyplot as plt


# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = tp.index
sizes = tp
  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots(figsize=(10,10))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
sol=df1[['Solved','Topic']]
sol.head()
```

```python
print(sol.shape)
```

```python

solved_dist=[]
for data in topics:
    dist={"Topic":str, "Solved":int,"Unsolved":int}
    solved_count=0
    for i in range(sol.shape[0]):
        if(data==sol['Topic'][i]):
            if sol['Solved'][i]==1:
                solved_count+=1
                
    dist["Topic"]=data
    dist["Solved"]=solved_count
    dist["Unsolved"]=tp[data]-solved_count
    solved_dist.append(dist)
        
        
```

```python
print(solved_dist)
```

```python
from pandas.io.json import json_normalize
df_sol = json_normalize(solved_dist)
print(df_sol)
```

```python
df_sol.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
N = df_sol.shape[0]
solved = df_sol["Solved"]
unsolved=df_sol["Unsolved"]
ind=df_sol["Topic"]
#ind = np.arange(N) # the x locations for the groups
width = 0.8
fig = plt.figure(figsize=(15,10))
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, solved, width, color='g')
ax.bar(ind, unsolved, width,bottom=solved, color='r')
ax.set_ylabel('Number of problems')
ax.set_title('Distribution of solved and unsolved questions as per topic.')
ax.set_xticks(ind )
ax.set_yticks(np.arange(0, 40, 1))
ax.legend(labels=['Solved', 'Unsolved'])
plt.show()
```
